## **操作方法**

プレイするときは必ず、**上のセルから順に動かしていってください。**  
最初にプレイするときは、チュートリアルを受けるのをお勧めします。

またゲーム中**「エンターを押してください。▼」と表記された場合**もエンターを押すことでゲームが進行する仕様になっているので**エンターを押してください。**
